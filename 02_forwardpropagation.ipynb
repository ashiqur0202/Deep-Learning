# Forwardpropagation


## Simple Network

```python
def relu(z):
    return max(0, z)

def feed_forward(x, Wh, Wo): # x is input, Wo and Wh are the weights 
    # Hidden layer
    Zh = x * Wh
    H = relu(Zh)
    
    # Output Layer
    Zo = H * Wo
    output = relu(Zo)
    return output
# Zo, Zh are weighted inputs

```

### Forward Propagation using Matrics

```python
import numpy as np

# Assuming these are global variables
input_layer_size = 1
hidden_layer_size = 2
output_layer_size = 2

# Initialize weights
def init_weights():
    Wh = np.random.randn(input_layer_size, hidden_layer_size) * np.sqrt(2.0 / input_layer_size)
    Wo = np.random.randn(hidden_layer_size, output_layer_size) * np.sqrt(2.0 / hidden_layer_size)
    return Wh, Wo

# Initialize biases
def init_bias():
    Bh = np.full((1, hidden_layer_size), 0.1)
    Bo = np.full((1, output_layer_size), 0.1)
    return Bh, Bo

# Rectified Linear Unit (ReLU) activation function
def relu(Z):
    return np.maximum(0, Z)

# Derivative of ReLU activation function
def relu_prime(Z):
    return np.where(Z > 0, 1, 0)

# Cost function
def cost(yHat, y):
    cost = np.sum((yHat - y)**2) / 2.0
    return cost

# Derivative of the cost function
def cost_prime(yHat, y):
    return yHat - y

# Feed-forward function
def feed_forward(X, Wh, Wo, Bh, Bo):
    # Hidden layer
    Zh = np.dot(X, Wh) + Bh
    H = relu(Zh)

    # Output layer
    Zo = np.dot(H, Wo) + Bo
    yHat = relu(Zo)

    return H, yHat

# Example usage
# Assuming X, Wh, Wo, Bh, Bo are defined elsewhere
X = np.array([[0.5]])  # Example input
Wh, Wo = init_weights()
Bh, Bo = init_bias()

# Forward propagation
H, yHat = feed_forward(X, Wh, Wo, Bh, Bo)
```

### Forward Propagation using Tensor

```python
import torch

# Assuming these are global variables
input_layer_size = 1
hidden_layer_size = 2
output_layer_size = 2

# Initialize weights
def init_weights():
    Wh = torch.randn((input_layer_size, hidden_layer_size)) * torch.sqrt(torch.tensor(2.0 / input_layer_size))
    Wo = torch.randn((hidden_layer_size, output_layer_size)) * torch.sqrt(torch.tensor(2.0 / hidden_layer_size))
    return Wh, Wo

# Initialize biases
def init_bias():
    Bh = torch.full((1, hidden_layer_size), 0.1)
    Bo = torch.full((1, output_layer_size), 0.1)
    return Bh, Bo

# Rectified Linear Unit (ReLU) activation function
def relu(Z):
    return torch.maximum(torch.tensor(0.0), Z)

# Derivative of ReLU activation function
def relu_prime(Z):
    return torch.where(Z > 0, torch.tensor(1.0), torch.tensor(0.0))

# Cost function
def cost(yHat, y):
    cost = torch.sum((yHat - y)**2) / 2.0
    return cost

# Derivative of the cost function
def cost_prime(yHat, y):
    return yHat - y

# Feed-forward function
def feed_forward(X, Wh, Wo, Bh, Bo):
    # Hidden layer
    Zh = torch.matmul(X, Wh) + Bh
    H = relu(Zh)

    # Output layer
    Zo = torch.matmul(H, Wo) + Bo
    yHat = relu(Zo)

    return H, yHat

# Example usage
# Assuming X, Wh, Wo, Bh, Bo are defined elsewhere
X = torch.tensor([[0.5]])  # Example input
Wh, Wo = init_weights()
Bh, Bo = init_bias()

# Forward propagation
H, yHat = feed_forward(X, Wh, Wo, Bh, Bo)
```
